DATA COLLECTING FROM PDF

```python
import camelot
import os

file_pdf = "data_06-02-2025_Laporan_Tahunan__Publikasi_Data_PMI_2024.pdf"
output_folder = "hasil_tabel"
os.makedirs(output_folder, exist_ok=True)

total_tabel = 0

for page in range(1, 54):
    tables = camelot.read_pdf(
        file_pdf,
        pages=str(page),
        flavor="stream"
    )
    if tables.n > 0:
        for i, t in enumerate(tables):
            file_name = os.path.join(output_folder, f"hal{page}_tabel{i+1}.csv")
            t.df.to_csv(file_name, index=False)
            print(f"Halaman {page}, tabel {i+1} disimpan.")
        total_tabel += tables.n

print(f"Total tabel berhasil diekstrak: {total_tabel}")
```

DATA INTEGRATION

```python
df_2022 = pd.read_excel("Cleaned_Pencari_Kerja_2022.xlsx")
df_2023 = pd.read_excel("Cleaned_Pencari_Kerja_2023.xlsx")

df_2022["Tahun"] = 2022
df_2023["Tahun"] = 2023

df_2022.rename(columns={"Nama Provinsi": "Provinsi"}, inplace=True)
df_2023.rename(columns={"Nama Provinsi": "Provinsi"}, inplace=True)

df_2022["Provinsi"] = df_2022["Provinsi"].str.strip()
df_2023["Provinsi"] = df_2023["Provinsi"].str.strip()

df_gabungan = pd.concat([df_2022, df_2023], ignore_index=True)
df_gabungan = df_gabungan.sort_values(by=["Tahun", "Provinsi"]).reset_index(drop=True)
df_gabungan.to_excel("Cleaned_Pencari_Kerja_2022_2023_Gabungan.xlsx", index=False)

df_gabungan.head()
```

```python
df_2022_2023 = pd.read_excel("Cleaned_Pencari_Kerja_2022_2023_Gabungan.xlsx")
df_2024 = pd.read_excel("Cleaned_Pencari_Kerja_2024.xlsx")

df_2024["Tahun"] = 2024

df_2022_2023.rename(columns={"Nama Provinsi": "Provinsi"}, inplace=True)
df_2024.rename(columns={"Nama Provinsi": "Provinsi"}, inplace=True)

df_2022_2023["Provinsi"] = df_2022_2023["Provinsi"].str.strip()
df_2024["Provinsi"] = df_2024["Provinsi"].str.strip()

df_gabungan2 = pd.concat([df_2022_2023, df_2024], ignore_index=True)
df_gabungan2 = df_gabungan2.sort_values(by=["Tahun", "Provinsi"]).reset_index(drop=True)
df_gabungan2.to_excel("Cleaned_Pencari_Kerja_2022_2023_2024_Gabungan.xlsx", index=False)

df_gabungan2.head()
```

```python
df_penempatan = pd.read_excel("Cleaned_Sorted_Transform 2_tabel_24_Penempatan Provinsi 2022-2024.xlsx")
df_pencari = pd.read_excel("Cleaned_Pencari_Kerja_2022_2023_2024_Gabungan.xlsx")

df_penempatan.rename(columns={'Nama Provinsi': 'Provinsi'}, inplace=True)
df_pencari.rename(columns={'Nama Provinsi': 'Provinsi'}, inplace=True)

df_penempatan['Provinsi'] = df_penempatan['Provinsi'].astype(str).str.strip()
df_pencari['Provinsi'] = df_pencari['Provinsi'].astype(str).str.strip()

df_penempatan['Tahun'] = pd.to_numeric(df_penempatan['Tahun'], errors='coerce')
df_pencari['Tahun'] = pd.to_numeric(df_pencari['Tahun'], errors='coerce')

Data_Gabungan_pekerja_Penempatan = pd.merge(df_pencari, df_penempatan, on=['Provinsi', 'Tahun'], how='inner')

merged = Data_Gabungan_pekerja_Penempatan.sort_values(by=['Tahun', 'Provinsi']).reset_index(drop=True)

cols = merged.columns.tolist()
if cols[0] != 'Tahun':
    cols.insert(0, cols.pop(cols.index('Tahun')))
    merged = merged[cols]

merged.to_excel("Data_gabungan_sorted_by_tahun.xlsx", index=False)

merged.head()
```

```python
df_Data_Gabungan_Sorted = pd.read_excel("Data_gabungan_sorted_by_tahun.xlsx")
df_Data_Gabungan_UMP = pd.read_excel("Cleaned_UMP_2022_2023_2024_Gabungan.xlsx")

df_Data_Gabungan_Sorted.rename(columns={'Nama Provinsi': 'Provinsi'}, inplace=True)
df_Data_Gabungan_UMP.rename(columns={'Nama Provinsi': 'Provinsi'}, inplace=True)

df_Data_Gabungan_Sorted['Provinsi'] = df_Data_Gabungan_Sorted['Provinsi'].astype(str).str.strip()
df_Data_Gabungan_UMP['Provinsi'] = df_Data_Gabungan_UMP['Provinsi'].astype(str).str.strip()

df_Data_Gabungan_Sorted['Tahun'] = pd.to_numeric(df_Data_Gabungan_Sorted['Tahun'], errors='coerce')
df_Data_Gabungan_UMP['Tahun'] = pd.to_numeric(df_Data_Gabungan_UMP['Tahun'], errors='coerce')

Data_Gabungan_pekerja_Penempatan_UMP = pd.merge(df_Data_Gabungan_UMP, df_Data_Gabungan_Sorted, on=['Provinsi', 'Tahun'], how='inner')

merged2 = Data_Gabungan_pekerja_Penempatan_UMP.sort_values(by=['Tahun', 'Provinsi']).reset_index(drop=True)

cols = merged2.columns.tolist()
if cols[0] != 'Tahun':
    cols.insert(0, cols.pop(cols.index('Tahun')))
    merged2 = merged2[cols]

merged2.to_excel("Data_Final.xlsx", index=False)

merged2
```
